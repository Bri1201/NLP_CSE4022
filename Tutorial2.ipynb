### 5.Homework Task

#### Determine the frequency distribution and conditional frequency distribution of any one of the presidential inaugural address

->POS tag each word and look at all the nouns, because they are objects and subjects that make a difference

->Nouns mean business

-> Histograms for frequency of words

```python
from nltk.corpus import inaugural
```

##### Let us look at Biden's speech 

```python
len(inaugural.paras(fileids= '2021-Biden.txt'))
```

```python
len(inaugural.sents(fileids = '2021-Biden.txt'))
```

```python
len(inaugural.words(fileids= '2021-Biden.txt'))
```

```python
Bidens_speech = ""

#Tags for nouns: "NN": Noun singular, "NNS": Noun plural, "NNP": Proper Noun, "NNPS": Proper noun plural
tag = nltk.pos_tag(inaugural.words(fileids='2021-Biden.txt'))
tag
for k in tag:
    #Not Considering proper nouns
    if( k[1] == "NN" or k[1] == "NNS"):
        Bidens_speech = Bidens_speech + " " + k[0]
    
```

#### Finding the word frequency

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fd = nltk.FreqDist(word.lower() for word in nltk.word_tokenize(Bidens_speech))
#The number of words are
fd.B()
```

```python
fd.plot()
```

Here we can see that the plot doesn't show us many insights as there are too many words
Another observation we can make is that majority of the words have occured only once.

Let us now include only the most common words, to start off we will include the most common 20

```python
fdNew = nltk.FreqDist(nltk.word_tokenize(Bidens_speech)).most_common(50)
```

```python


## Conversion to Pandas series via Python Dictionary for easier plotting
all_fdist = pd.Series(dict(fdNew))

## Setting figure, ax into variables
fig, ax = plt.subplots(figsize=(10,10))

## Seaborn plotting using Pandas attributes + xtick rotation for ease of viewing
all_plot = sns.barplot(y=all_fdist.index, x=all_fdist.values, ax=ax, orient='h')
plt.title("Word Count")
plt.xlabel("Frequency")
plt.ylabel("Words")
```
